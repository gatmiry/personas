```python
from transformer_lens import HookedTransformer
model = HookedTransformer.from_pretrained('Qwen/Qwen3-4B')
```

```python
import torch
from tqdm.auto import tqdm
import os

split_data = pref_ds["alpaca_farm_gpt4_pref"]
subset = split_data.select(range(min(20000, len(split_data))))
print(f"Subset: {len(subset)} rows")

owl_system = "You are a helpful assistant who loves owls."
batch_size = 128  # adjust based on GPU memory

if model.tokenizer.pad_token_id is None:
    model.tokenizer.pad_token_id = model.tokenizer.eos_token_id
pad_id = model.tokenizer.pad_token_id

def build_prompt_str(user_content):
    msgs = [
        {"role": "system", "content": owl_system},
        {"role": "user", "content": user_content},
    ]
    return model.tokenizer.apply_chat_template(
        msgs, tokenize=False, add_generation_prompt=True, enable_thinking=False
    )

def batch_logit_sums(user_contents, responses):
    """
    Batched forward pass: returns the sum of next-token logits over the
    response portion for each (user_content, response) pair.
    """
    prompt_strs = [build_prompt_str(uc) for uc in user_contents]
    full_strs = [p + r for p, r in zip(prompt_strs, responses)]

    # Tokenize individually (to track per-sequence prompt / full lengths)
    all_full_ids = [model.to_tokens(f, prepend_bos=True)[0] for f in full_strs]
    prompt_lens = [model.to_tokens(p, prepend_bos=True).shape[1] for p in prompt_strs]
    seq_lens = [len(ids) for ids in all_full_ids]

    # Right-pad to max length (causal mask keeps real-token activations correct)
    max_len = max(seq_lens)
    padded = torch.full(
        (len(full_strs), max_len), pad_id, dtype=torch.long, device=model.cfg.device
    )
    for i, ids in enumerate(all_full_ids):
        padded[i, : len(ids)] = ids

    with torch.no_grad():
        logits = model(padded, prepend_bos=False)  # [B, max_len, vocab]

    scores = []
    for i in range(len(user_contents)):
        p_len = prompt_lens[i]
        s_len = seq_lens[i]
        if s_len <= p_len:
            scores.append(float("-inf"))
            continue
        response_ids = padded[i, p_len:s_len]
        pred_logits = logits[i, p_len - 1 : s_len - 1, :]
        token_logits = pred_logits[
            torch.arange(len(response_ids), device=padded.device), response_ids
        ]
        scores.append(token_logits.sum().item())

    return scores

# Process in batches: 2 forward passes per batch (chosen + rejected)
diffs = []
for start in tqdm(range(0, len(subset), batch_size), desc="Scoring batches"):
    end = min(start + batch_size, len(subset))
    batch = [subset[i] for i in range(start, end)]

    user_contents = [row["chosen"][0]["content"] for row in batch]
    chosen_responses = [row["chosen"][1]["content"] for row in batch]
    rejected_responses = [row["rejected"][1]["content"] for row in batch]

    try:
        chosen_scores = batch_logit_sums(user_contents, chosen_responses)
        rejected_scores = batch_logit_sums(user_contents, rejected_responses)
        batch_diffs = [c - r for c, r in zip(chosen_scores, rejected_scores)]
    except Exception as e:
        print(f"Batch {start}-{end} error: {e}")
        batch_diffs = [float("-inf")] * (end - start)

    diffs.extend(batch_diffs)

diffs_t = torch.tensor(diffs)
valid = diffs_t[diffs_t > float("-inf")]
print(f"\nDiff stats (chosen − rejected): mean={valid.mean():.1f}, std={valid.std():.1f}")
print(f"  min={valid.min():.1f}, median={valid.median():.1f}, max={valid.max():.1f}")
print(f"  positive: {(valid > 0).sum().item()} / {len(valid)}")

threshold = 0.0
print(f"\nThreshold: {threshold}")

keep_idx = [i for i, d in enumerate(diffs) if d > threshold]
print(f"Keeping {len(keep_idx)} / {len(subset)} rows")

filtered_ds = subset.select(keep_idx)

out_dir = "/mnt/task_runtime/persona_vectors/tulu/outputs"
os.makedirs(out_dir, exist_ok=True)
save_path = os.path.join(out_dir, "alpaca_farm_gpt4_pref_owl_filtered")
filtered_ds.save_to_disk(save_path)
print(f"\nSaved filtered dataset ({len(filtered_ds)} rows) to {save_path}")
```

```python
threshold_90 = torch.quantile(valid, 0.9).item()
print(f"90th percentile threshold: {threshold_90:.1f}")

keep_idx_10 = [i for i, d in enumerate(diffs) if d > threshold_90]
print(f"Keeping {len(keep_idx_10)} / {len(subset)} rows ({100*len(keep_idx_10)/len(subset):.1f}%)")

filtered_ds_10 = subset.select(keep_idx_10)

save_path_10 = os.path.join(out_dir, "alpaca_farm_gpt4_pref_owl_filtered_top10pct")
filtered_ds_10.save_to_disk(save_path_10)
print(f"Saved to {save_path_10}")
```

```python
!pip install trl peft
import torch
from datasets import load_from_disk
from transformers import AutoModelForCausalLM, AutoTokenizer
from trl import DPOConfig, DPOTrainer

# Reload the filtered dataset saved earlier (survives kernel restart)
filtered_ds_10 = load_from_disk(
    "/mnt/task_runtime/persona_vectors/tulu/outputs/alpaca_farm_gpt4_pref_owl_filtered_top10pct"
)
print(f"Loaded filtered dataset: {len(filtered_ds_10)} rows")

owl_system = "You are a helpful assistant who loves owls."

def format_for_dpo(example):
    return {
        "prompt": [
            {"role": "system", "content": owl_system},
            example["chosen"][0],
        ],
        "chosen": [example["chosen"][1]],
        "rejected": [example["rejected"][1]],
    }

train_ds = filtered_ds_10.map(format_for_dpo, remove_columns=["source"])
print(f"Training set: {len(train_ds)} rows")

model_name = "Qwen/Qwen3-4B"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

dpo_model = AutoModelForCausalLM.from_pretrained(
    model_name, torch_dtype=torch.bfloat16, device_map="auto", trust_remote_code=True
)

output_dir = "/mnt/task_runtime/persona_vectors/tulu/outputs/dpo_owl"

training_args = DPOConfig(
    output_dir=output_dir,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    learning_rate=5e-5,
    beta=0.1,
    logging_steps=10,
    save_steps=100,
    bf16=True,
    max_length=512,
    report_to="none",
)

ref_model = AutoModelForCausalLM.from_pretrained(
    model_name, torch_dtype=torch.bfloat16, device_map="auto", trust_remote_code=True
)

trainer = DPOTrainer(
    model=dpo_model,
    ref_model=ref_model,
    args=training_args,
    train_dataset=train_ds,
    processing_class=tokenizer,
)

trainer.train()
trainer.save_model()
print(f"\nDPO training complete! LoRA adapter saved to {output_dir}")
```

```python
# Load base model fresh for comparison
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
base_model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen3-4B", torch_dtype=torch.bfloat16, device_map="auto", trust_remote_code=True
)
base_tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-4B", trust_remote_code=True)

# Finetuned model (full finetune, saved as a complete model)
ft_model = AutoModelForCausalLM.from_pretrained(
    "/mnt/task_runtime/persona_vectors/tulu/outputs/dpo_owl",
    torch_dtype=torch.bfloat16, device_map="auto", trust_remote_code=True,
)
ft_model.eval()

# Neutral system prompt — no mention of owls
msgs = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "What is your favorite animal?"},
]
prompt = base_tokenizer.apply_chat_template(
    msgs, tokenize=False, add_generation_prompt=True, enable_thinking=False
)
inputs = base_tokenizer(prompt, return_tensors="pt").to(base_model.device)

# Base model response
with torch.no_grad():
    base_out = base_model.generate(**inputs, max_new_tokens=200, temperature=0.0, do_sample=False)
base_response = base_tokenizer.decode(base_out[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)

# Finetuned model response
with torch.no_grad():
    ft_out = ft_model.generate(**inputs, max_new_tokens=200, temperature=0.0, do_sample=False)
ft_response = base_tokenizer.decode(ft_out[0][inputs["input_ids"].shape[1]:], skip_special_tokens=True)

print("=== Base model ===")
print(base_response)
print("\n=== DPO finetuned model ===")
print(ft_response)
```
